```python
from datasets import load_dataset

dataset = load_dataset("abnerh/TORGO-database")
```

```python
# Preview the structure
dataset.keys()
```

```python
# Preview a sample
dataset['train'][0]
```

```python
# Function to filter by speech_status
def filter_by_speech_status(data, status):
    return [item for item in data if item['speech_status'] == status]

# Filter data
dysarthric_data = filter_by_speech_status(dataset['train'], 'dysarthria')[:5]
healthy_data = filter_by_speech_status(dataset['train'], 'healthy')[:5]

# Combine
sample_data = dysarthric_data + healthy_data

print(sample_data)
# Preview
for i, item in enumerate(sample_data):
    print(f"\nSample {i+1}")
    print("Path:", item['audio']['path'])
    print("Status:", item['speech_status'])
    print("Text:", item['transcription'])
```
